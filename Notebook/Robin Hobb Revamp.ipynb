```python
import matplotlib.pyplot as plt
import numpy as np

robinhobb_dollars = RobinHobb[['Title', 'YTD Dollars', '2023 Dollars', '2022 Dollars', '2021 Dollars', '2020 Dollars']]
robinhobb_units = RobinHobb[['Title', 'YTD Sales', '2023', '2022', '2021', '2020']]

robinhobb_dollars = robinhobb_dollars[['Title', '2020 Dollars', '2021 Dollars', '2022 Dollars', '2023 Dollars', 'YTD Dollars']]
robinhobb_units = robinhobb_units[['Title', '2020', '2021', '2022', '2023', 'YTD Sales']]

fig, ax1 = plt.subplots(figsize=(10, 8))

bar_width = 0.15
x = np.arange(len(robinhobb_dollars['Title']))

colors = ['rosybrown', 'indianred', 'brown', 'sienna', 'maroon']

for i, year in enumerate(robinhobb_dollars.columns[1:]):
    ax1.bar(x + i * bar_width, robinhobb_dollars[year], bar_width, label=year, color=colors[i], alpha=0.6)

ax1.set_xlabel('Title', fontweight='bold')
ax1.set_ylabel('Total Dollars ($)', fontweight='bold')
ax1.set_title('Total Dollars and Units Over the Years for The Rain Wild Chronicles', fontweight='bold')
ax1.set_xticks(x + 2 * bar_width)
ax1.set_xticklabels(robinhobb_dollars['Title'], fontweight='bold')

ax2 = ax1.twinx()

for i, year in enumerate(robinhobb_units.columns[1:]):
    ax2.plot(x + i * bar_width, robinhobb_units[year], marker='o', label=f'Units {year}', color=colors[i], linewidth=2)

ax2.set_ylabel('Total Units', fontweight='bold')

handles, labels = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()

legend = fig.legend(handles[:5] + handles2[:5], ['2020', '2021', '2022', '2023', 'YTD'], loc='upper center', bbox_to_anchor=(0.5, -.001), ncol=5, frameon=True, title='Years')

frame = legend.get_frame()
frame.set_edgecolor('black')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
RobinHobbWeekly.columns = RobinHobbWeekly.columns.str.strip()
RobinHobbWeekly.columns = ['Week', 'City of Dragons', 'Dragon Keeper',
       'Dragon Haven', 'Blood of Dragons']
```

```python
RobinHobbWeekly['Start Date'] = pd.to_datetime(RobinHobbWeekly['Week'].str.split(' - ').str[0], format='%b %d %Y')
start_date = pd.to_datetime('Dec 30 2023')
filtered_df = RobinHobbWeekly[RobinHobbWeekly['Start Date'] >= start_date]
filtered_df = RobinHobbWeekly[RobinHobbWeekly['Start Date'] >= start_date]
filtered_df = filtered_df.sort_values(by='Start Date')
filtered_df = filtered_df.drop(columns=['Start Date'])
filtered_df.reset_index(drop=True, inplace=True)
filtered_df
```

```python
filtered_df['Dragon Keeper'] = pd.to_numeric(filtered_df['Dragon Keeper'], errors='coerce')
filtered_df['Dragon Haven'] = pd.to_numeric(filtered_df['Dragon Haven'], errors='coerce')
filtered_df['City of Dragons'] = pd.to_numeric(filtered_df['City of Dragons'], errors='coerce')
filtered_df['Blood of Dragons'] = pd.to_numeric(filtered_df['Blood of Dragons'], errors='coerce')

avg_weekly_sales = filtered_df[['Dragon Keeper', 'Dragon Haven', 'City of Dragons', 'Blood of Dragons']].mean()
avg_weekly_sales
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(avg_weekly_sales.index, avg_weekly_sales.values, color=['moccasin', 'papayawhip', 'sandybrown', 'lightseagreen'])

plt.title('YTD Average Weekly Sales for The Rain Wild Chronicles', fontweight='bold')
plt.xlabel('Book Titles', fontweight='bold')
plt.ylabel('Average Weekly Sales', fontweight='bold')
plt.xticks()
plt.grid(axis='y', linestyle='--', alpha=0.7)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', fontweight='bold')
plt.tight_layout()
plt.show()
```

The relatively stable sales for the final books suggest that those who enjoy the series tend to stay with it until the end.

```python
farseer_titles = [
    "Assassin's Apprentice: the Farseer Trilogy Book 1", 
    "Royal Assassin", 
    "Assassin's Quest: the Farseer Trilogy Book 3"
]
liveship_titles = [
    "Ship of Magic", 
    "Mad Ship", 
    "Ship of Destiny"
]
tawny_man_titles = [
    "Fool's Errand: the Tawny Man Trilogy Book 1", 
    "Golden Fool: the Tawny Man Trilogy Book 2", 
    "Fool's Fate: the Tawny Man Trilogy Book III"
]
rain_wild_titles = [
    "Dragon Keeper", 
    "Dragon Haven", 
    "City of Dragons", 
    "Blood of Dragons"
]
fitz_and_fool_titles = [
    "Fool's Assassin", 
    "Fool's Quest", 
    "Assassin's Fate"
]

farseer_df = rote[rote['Title'].isin(farseer_titles)]
liveship_df = rote[rote['Title'].isin(liveship_titles)]
tawny_man_df = rote[rote['Title'].isin(tawny_man_titles)]
rain_wild_df = rote[rote['Title'].isin(rain_wild_titles)]
fitz_and_fool_df = rote[rote['Title'].isin(fitz_and_fool_titles)]
```

```python
farseer_sum = farseer_df[['YTD Sales', '2023', '2022', '2021', '2020']].apply(pd.to_numeric, errors='coerce').sum().sum()
liveship_sum = liveship_df[['YTD Sales', '2023', '2022', '2021', '2020']].apply(pd.to_numeric, errors='coerce').sum().sum()
tawny_man_sum = tawny_man_df[['YTD Sales', '2023', '2022', '2021', '2020']].apply(pd.to_numeric, errors='coerce').sum().sum()
rain_wild_sum = rain_wild_df[['YTD Sales', '2023', '2022', '2021', '2020']].apply(pd.to_numeric, errors='coerce').sum().sum()
fitz_and_fool_sum = fitz_and_fool_df[['YTD Sales', '2023', '2022', '2021', '2020']].apply(pd.to_numeric, errors='coerce').sum().sum()
```

```python
import matplotlib.pyplot as plt
sales_data = {
    'Series': ['The Farseer Trilogy', 'Liveship Traders Trilogy', 'The Tawny Man Trilogy', 'The Rain Wild Chronicles', 'The Fitz and the Fool Trilogy'],
    'Total Units': [farseer_sum, liveship_sum, tawny_man_sum, rain_wild_sum, fitz_and_fool_sum]
}

sales_df = pd.DataFrame(sales_data)

labels = ['The Farseer Trilogy', 'Liveship Traders Trilogy', 'The Tawny Man Trilogy', r'$\bf{The\ Rain\ Wild\ Chronicles}$', 'The Fitz and the Fool Trilogy']

explode = [0, 0, 0, 0.1, 0]

colors = ['darkseagreen', 'darksalmon', 'khaki', 'lightblue', 'thistle']

fig, ax = plt.subplots()
ax.pie(sales_df['Total Units'], labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, explode=explode,
       textprops={'color': 'black'})  
for text in ax.texts:
    if text.get_text() in labels:
        text.set_color('white')
        text.set_fontweight('bold')

ax.axis('equal')
fig.patch.set_alpha(0.0)
ax.patch.set_alpha(0.0)
plt.show()
```
